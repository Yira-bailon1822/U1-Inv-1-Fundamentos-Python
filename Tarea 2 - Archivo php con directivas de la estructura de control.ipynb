  Reto para el alumno: Crea una función que calcule el precio de un producto aplicando el IVA (16%). Debe recibir el precio base y retornar el total.

```python
# Definición de la función con parámetros
def calcular_precio_con_iva(precio_base: float) -> float:
    iva = 0.16
    total = precio_base + (precio_base * iva)
    return total

# Pruebas de la función
producto_A = 150.0
producto_B = 499.99

print(f"Total Producto A con IVA: ${calcular_precio_con_iva(producto_A):.2f}")
print(f"Total Producto B con IVA: ${calcular_precio_con_iva(producto_B):.2f}")
```

# Métodos HTTP

**La analogía de un Restaurante:**

  El Cliente (Navegador/Postman): El comensal que se sienta a la mesa.

  Los Métodos HTTP: La acción que el cliente quiere realizar (Ver el menú, pedir comida, cambiar su orden o cancelar un plato).

  FastAPI: El mesero que toma la orden, la lleva a la cocina (las funciones de Python) y regresa con un resultado y un estado (si el plato está listo o si se acabó el ingrediente).

**1. Los Métodos HTTP (Verbos)**

Para gestionar un recurso (por ejemplo, los datos de un Alumno), usamos cuatro verbos principales:

*    GET (Leer): El cliente solicita información. No modifica nada en el servidor. (Equivale a pedir ver la carta del restaurante).

 *   POST (Crear): El cliente envía datos nuevos para que el servidor los guarde. (Equivale a ordenar un plato nuevo).

 *   PUT (Actualizar): El cliente envía datos para modificar o reemplazar un recurso existente. (Equivale a cambiar tu plato de sopa por uno de carne).

*    DELETE (Borrar): El cliente solicita eliminar un recurso. (Equivale a cancelar un pedido antes de que lo cocinen).




# Códigos de Estado

**2. Los Códigos de Estado (Status Codes)**

El servidor siempre debe responder con un número de tres dígitos que le indica al cliente qué pasó con su petición. Se dividen en familias fáciles de recordar:

 *   Familia 200 (Éxito): "¡Todo salió bien!"

      * 200 OK: La petición fue exitosa (común en GET y PUT).
      * 201 Created: El recurso se creó con éxito (común en POST).

 *   Familia 300 (Redirección): "El recurso se movió de lugar".

      * 301 Moved Permanently: La página o API ahora está en otra dirección web.

 *   Familia 400 (Errores del Cliente): "Es culpa del usuario".

      * 400 Bad Request: El cliente mandó datos mal estructurados.

      * 401 Unauthorized: El usuario no ha iniciado sesión o no tiene permiso.

      * 404 Not Found: Lo que el cliente busca no existe (el clásico error web).

 *   Familia 500 (Errores del Servidor): "Es culpa del programador / sistema".

      * 500 Internal Server Error: El código de Python falló (hubo un crash o bug) y el servidor no supo qué hacer.

# CRUD Ejemplo con FastAPI

**Un CRUD Completo con Respuestas Correctas**

Tus alumnos pueden copiar y ejecutar este código directamente en Colab. Simularemos una base de datos usando una lista de Python para que vean cómo interactúan las clases (POO), las funciones y los métodos HTTP.

```python
# -------------------------------------------------------------------------
# 1. INSTALACIÓN Y CONFIGURACIÓN DEL ENTORNO
# -------------------------------------------------------------------------

# Instala las librerías necesarias en el servidor de Google Colab de forma silenciosa (--quiet)
!pip install fastapi uvicorn nest_asyncio requests --quiet

# Permite que el servidor web corra dentro de un entorno de bucles anidados como el de Colab
import nest_asyncio
# El núcleo del framework para crear nuestra aplicación web (API)
from fastapi import FastAPI, HTTPException, status
# Herramienta para obligar a que los datos cumplan con una estructura estricta (POO)
from pydantic import BaseModel
# Nos permite avisarle a Python que vamos a manejar colecciones (listas) de objetos específicos
from typing import List
# El servidor web que se encarga de escuchar y responder las peticiones HTTP externas
import uvicorn
# Permite encender el servidor web en un "hilo" separado para no congelar la pantalla de Colab
import threading

# Aplica el parche necesario para que Uvicorn no choque con el sistema interno de Colab
nest_asyncio.apply()

# Creamos la instancia principal de la aplicación web y le asignamos un título descriptivo
app = FastAPI(title="API Sistema de Alumnos")


# -------------------------------------------------------------------------
# 2. MODELO DE DATOS (POO CON PYDANTIC)
# -------------------------------------------------------------------------

# Creamos una clase que hereda de BaseModel. Esto define el "molde" exacto que usará cada Alumno
class Alumno(BaseModel):
    id: int           # Obliga a que el ID sea un número entero
    nombre: str       # Obliga a que el nombre sea una cadena de texto
    carrera: str      # Obliga a que la carrera sea una cadena de texto
    promedio: float   # Obliga a que el promedio sea un número con decimales

# Creamos una lista normal de Python que simulará ser nuestra Base de Datos en la memoria RAM.
# Declaramos que es una lista ("List") que adentro solo puede guardar objetos de tipo "Alumno".
base_de_datos_alumnos: List[Alumno] = [
    # Instanciamos los dos primeros registros de prueba usando nuestro molde
    Alumno(id=1, nombre="Juan Pérez", carrera="Sistemas", promedio=85.0),
    Alumno(id=2, nombre="Maria Lopez", carrera="Industrial", promedio=92.5)
]


# -------------------------------------------------------------------------
# 3. ENDPOINTS O RUTAS DE LA API (MÉTODOS HTTP)
# -------------------------------------------------------------------------

# --- MÉTODO GET: OBTENER RECURSOS ---
# Registra una ruta web en la dirección "/alumnos". Si sale bien, responderá con código 200 OK
@app.get("/alumnos", status_code=status.HTTP_200_OK)
def obtener_alumnos():
    # El código de Python simplemente retorna la lista completa y FastAPI la convierte a JSON
    return base_de_datos_alumnos


# --- MÉTODO POST: CREAR RECURSOS ---
# Registra una ruta tipo POST. Al crear un registro con éxito, responderá con código 201 Created
@app.post("/alumnos", status_code=status.HTTP_201_CREATED)
def crear_alumno(nuevo_alumno: Alumno): # FastAPI lee el JSON entrante y lo convierte en el objeto "nuevo_alumno"

    # Recorremos nuestra lista para revisar si ya existe un alumno con ese mismo ID
    for alumno in base_de_datos_alumnos:
        if alumno.id == nuevo_alumno.id:
            # Si el ID ya existe, detenemos todo lanzando una excepción HTTP con código 400 (Error del cliente)
            raise HTTPException(
                status_code=status.HTTP_400_BAD_REQUEST,
                detail="El ID de este alumno ya está registrado."
            )

    # Si el bucle termina y el ID es único, agregamos el nuevo objeto a nuestra lista en memoria
    base_de_datos_alumnos.append(nuevo_alumno)
    # Retornamos un mensaje de confirmación y los datos que acabamos de guardar
    return {"mensaje": "Alumno registrado con éxito", "alumno": nuevo_alumno}


# --- MÉTODO PUT: ACTUALIZAR RECURSOS ---
# El parámetro {alumno_id} en la URL nos permite saber exactamente a quién queremos modificar
@app.put("/alumnos/{alumno_id}", status_code=status.HTTP_200_OK)
def actualizar_alumno(alumno_id: int, datos_actualizados: Alumno):

    # "enumerate" nos da tanto la posición en la lista (indice) como el objeto (alumno)
    for indice, alumno in enumerate(base_de_datos_alumnos):
        # Si encontramos al alumno cuyo ID coincide con el solicitado en la URL...
        if alumno.id == alumno_id:
            # Reemplazamos el objeto viejo en esa posición por el nuevo molde que llegó modificado
            base_de_datos_alumnos[indice] = datos_actualizados
            # Respondemos avisando que todo salió bien
            return {"mensaje": "Alumno actualizado con éxito", "alumno": datos_actualizados}

    # Si el bucle termina y nadie coincidió con el ID, lanzamos un Error 404 (No encontrado)
    raise HTTPException(
        status_code=status.HTTP_404_NOT_FOUND,
        detail=f"Alumno con ID {alumno_id} no fue encontrado."
    )


# --- MÉTODO DELETE: ELIMINAR RECURSOS ---
# Solicitamos el ID del alumno que se desea dar de baja a través de la URL
@app.delete("/alumnos/{alumno_id}", status_code=status.HTTP_200_OK)
def eliminar_alumno(alumno_id: int):

    # Recorremos la lista buscando al alumno
    for indice, alumno in enumerate(base_de_datos_alumnos):
        if alumno.id == alumno_id:
            # ".pop(indice)" saca al elemento de la lista definitivamente y lo guarda un momento en la variable
            alumno_eliminado = base_de_datos_alumnos.pop(indice)
            # Despedimos al alumno confirmando su nombre en la respuesta
            return {"mensaje": f"El alumno {alumno_eliminado.nombre} fue eliminado."}

    # Si el ID a eliminar no existía en el sistema, respondemos con un Error 404
    raise HTTPException(
        status_code=status.HTTP_404_NOT_FOUND,
        detail="No se pudo eliminar. ID no encontrado."
    )


# -------------------------------------------------------------------------
# 4. LANZAMIENTO DEL SERVIDOR WEB
# -------------------------------------------------------------------------

# Definimos una función estándar de Python dedicada exclusivamente a encender el motor web
def encender_servidor():
    # Configura Uvicorn para que corra nuestra app en la IP local y en el puerto 8000
    uvicorn.run(app, host="127.0.0.1", port=8000)

# Iniciamos un hilo (thread) secundario apuntando a nuestra función para que el servidor corra "de fondo"
threading.Thread(target=encender_servidor, daemon=True).start()

# Mensaje final en la consola de Colab indicando que el entorno está listo para recibir peticiones
print("Servidor listo y escuchando peticiones.")
```

**Validacion de endpoint de FastAPI**

```python
# Importamos la librería estándar para simular una aplicación externa (como Postman o el navegador)
import requests

# Guardamos la dirección base de nuestra API en una constante para no escribirla a cada rato
URL = "http://127.0.0.1:8000/alumnos"

print("=== 1. PROBANDO GET (Obtener Lista Original) ===")
# Simula una petición HTTP GET a la URL de nuestra API
res = requests.get(URL)
# ".status_code" nos muestra el número de tres dígitos que devolvió la API (Debería ser 200)
print(f"Código de Estado: {res.status_code}")
# ".json()" transforma la respuesta de texto plano a un formato legible por Python
print(res.json(), "\n")


print("=== 2. PROBANDO POST (Crear Alumno Nuevo) ===")
# Creamos un diccionario con la estructura exacta que exige nuestro molde de Pydantic
nuevo = {"id": 3, "nombre": "Carlos Gómez", "carrera": "Mecatrónica", "promedio": 88.0}
# Enviamos la petición POST adjuntando el diccionario dentro del parámetro "json"
res = requests.post(URL, json=nuevo)
print(f"Código de Estado: {res.status_code}") # Debería devolver 201 (Created)
print(res.json(), "\n")


print("=== 3. PROBANDO CONTROL DE ERRORES (Provocar un 400 Bad Request) ===")
# Volvemos a enviar los mismos datos. Al repetir el ID 3, se activará el "if" que programamos arriba
res = requests.post(URL, json=nuevo)
print(f"Código de Estado: {res.status_code}") # Debería devolver 400 (Bad Request)
print(res.json(), "\n")


print("=== 4. PROBANDO PUT (Modificar al Alumno con ID 1) ===")
# Preparamos los datos modificados
cambio = {"id": 1, "nombre": "Juan Pérez Modificado", "carrera": "Sistemas Avanzados", "promedio": 95.0}
# Concatenamos el ID "1" al final de la URL para indicarle a la API a quién actualizar (/alumnos/1)
res = requests.put(f"{URL}/1", json=cambio)
print(f"Código de Estado: {res.status_code}") # Debería devolver 200 (OK)
print(res.json(), "\n")


print("=== 5. PROBANDO DELETE (Eliminar al Alumno con ID 2) ===")
# Solicitamos la eliminación del alumno mandando el ID "2" al final de la URL (/alumnos/2)
res = requests.delete(f"{URL}/2")
print(f"Código de Estado: {res.status_code}") # Debería devolver 200 (OK)
print(res.json(), "\n")


print("=== 6. PROBANDO ERROR 404 (Buscar un ID inexistente) ===")
# Intentamos borrar de nuevo al alumno 2. Como ya fue eliminado, la API no lo encontrará
res = requests.delete(f"{URL}/2")
print(f"Código de Estado: {res.status_code}") # Debería devolver 404 (Not Found)
print(res.json(), "\n")
```
